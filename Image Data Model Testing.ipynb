```python
## Imports
import os
import shutil
import uuid
import cv2
import warnings
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

## Tensorflow Imports
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing import image

warnings.filterwarnings("ignore")
```

## Methods

```python
# import the required model
def import_model(model_path):
    model = keras.models.load_model(model_path)
    return model
```

```python
def predict_image(image_path, model):
    
    # dimensions of our images
    img_width, img_height = 224, 224

    # load image
    img = image.load_img(image_path, target_size=(img_width, img_height))
    
    # change image numpy array
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    img_arr = np.vstack([x])
    
    # predict
    prediction = model.predict(img_arr)
    pred = np.argmax(prediction, axis=1)
    
    #print(pred)
    
    plt.imshow(img)
    
    if pred == [1]:
        print("Model Prediction: Positive")
    elif pred == [0]:
        print("Model Prediction: Negative")
    else:
        print("Problem generating prediction!!")
        
    # check if model made the correct prediction
    if image_path is neg_example:
        
        # prediction should be Negative
        if pred != [0]:
            print("Model made incorrect prediction!")
        else: 
            print("Model made correct prediction!")
        
    elif image_path is pos_example:
        
        # prediction should be Negative
        if pred != [1]:
            print("Model made incorrect prediction!")
        else: 
            print("Model made correct prediction!")
            
    else: 
        print("Can't find image path!")
            
```
