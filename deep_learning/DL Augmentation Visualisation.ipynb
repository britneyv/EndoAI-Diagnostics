# DL Augmentation Visualisation

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#image paths
neg_example = 'C:/Users/vella/Mastes Thesis/processed_data/no_pathology_train/f_0.jpg6cc12c80-c86c-4813-9971-280492b00017'
pos_example = 'C:/Users/vella/Mastes Thesis/processed_data/pathology_train/f_0_a.jpg0be0bd50-1173-4df0-bf2d-c9d82faf602a'
```

### Non-Pathology Example

```python
# Load and preprocess the image
img = image.load_img(neg_example)  # Load image
img_array = image.img_to_array(img)  # Convert to array
img_array = np.expand_dims(img_array, axis=0)  # Expand dimensions for batch processing

# Define individual augmentation generators
augmentations = {
    "Original Image": ImageDataGenerator(),
    "Rotation": ImageDataGenerator(rotation_range=30, fill_mode='nearest'),
    "Width Shift": ImageDataGenerator(width_shift_range=0.2,  fill_mode='nearest'),
    "Height Shift": ImageDataGenerator(height_shift_range=0.2,  fill_mode='nearest'),
    "Shearness": ImageDataGenerator(shear_range=0.2,  fill_mode='nearest'),
    "Zoom": ImageDataGenerator(zoom_range=0.2,  fill_mode='nearest'),
    "Horizontal Flip": ImageDataGenerator(horizontal_flip=True,  fill_mode='nearest'),
    "Brightness": ImageDataGenerator(brightness_range=[0.8, 1.2],  fill_mode='nearest'),
}

# Create figure and axes for a 3x3 grid
fig, axes = plt.subplots(2, 4, figsize=(12, 4))
axes = axes.ravel()

# Iterate through augmentations and display images
for i, (title, gen) in enumerate(augmentations.items()):
    if i == 0:
        augmented_image = img_array[0].astype('uint8')  # Original image
    else:
        aug_iter = gen.flow(img_array, batch_size=1)
        augmented_image = aug_iter.next()[0].astype('uint8')
    
    axes[i].imshow(augmented_image)
    axes[i].axis('off')
    axes[i].set_title(title, fontsize=14, color='darkblue')

plt.tight_layout()
plt.show()
```
