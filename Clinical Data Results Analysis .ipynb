```python
import pandas as pd
```

```python
model_results = pd.read_csv('model_results.csv', index_col=0)
model_results.head()
```

```python
# Set 'Metrics' column as index
model_results.set_index('Metrics', inplace=True)

# Calculate the average score for each metric
average_scores = model_results.mean(axis=1)

# Print results
print("Average Scores for Each Metric:")
print(average_scores)

# Find the highest and lowest scores for each metric
highest_scores = model_results.idxmax(axis=1)  # Model names with the highest scores
lowest_scores = model_results.idxmin(axis=1)  # Model names with the lowest scores

highest_values = model_results.max(axis=1)  # Highest score values for each metric
lowest_values = model_results.min(axis=1)  # Lowest score values for each metric

# Combine results into a summary
result = pd.DataFrame({
    'Metric': model_results.index,
    'Highest Model': highest_scores,
    'Highest Score': highest_values,
    'Lowest Model': lowest_scores,
    'Lowest Score': lowest_values
})

# Display results
result.head(10)
```

```python
# Number of top models to display
n = 3

# Find top 'n' models with highest and lowest scores for each metric
highest_scores = model_results.apply(lambda x: x.nlargest(n).index.tolist(), axis=1)
lowest_scores = model_results.apply(lambda x: x.nsmallest(n).index.tolist(), axis=1)

highest_values = model_results.apply(lambda x: x.nlargest(n).values.tolist(), axis=1)
lowest_values = model_results.apply(lambda x: x.nsmallest(n).values.tolist(), axis=1)

# Create a summary DataFrame for top 'n' models
summary = pd.DataFrame({
    'Metric': model_results.index,
    'Top Models (Highest Scores)': highest_scores,
    'Top Scores': highest_values,
    'Bottom Models (Lowest Scores)': lowest_scores,
    'Bottom Scores': lowest_values
})

# Display the summary
summary.head(10)
```

```python
# Number of top/bottom models to display
n = 3

# Function to get a tidy list for top and bottom scores
def get_tidy_list(df, n):
    tidy_list = []
    for metric in df.index:
        # Top n models for each metric
        top_models = df.loc[metric].nlargest(n).index
        top_scores = df.loc[metric].nlargest(n).values
        for i, (model, score) in enumerate(zip(top_models, top_scores), start=1):
            tidy_list.append({'Metric': metric, 'Placement': i, 'Model': model, 'Score': score, 'Type': 'Top'})

        # Bottom n models for each metric
        bottom_models = df.loc[metric].nsmallest(n).index
        bottom_scores = df.loc[metric].nsmallest(n).values
        for i, (model, score) in enumerate(zip(bottom_models, bottom_scores), start=1):
            tidy_list.append({'Metric': metric, 'Placement': i, 'Model': model, 'Score': score, 'Type': 'Bottom'})

    return pd.DataFrame(tidy_list)

# Get the tidy list
tidy_results = get_tidy_list(model_results, n)

# Sort the tidy list by metric and placement
tidy_results.sort_values(by=['Metric', 'Type', 'Placement'], inplace=True)

# Display the tidy results
tidy_results.head(50)
```
