## Image Augmentation Visualisation

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# different augmentations that will be randomly used by the ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=30,  # Randomly rotate images up to 30 degrees
    width_shift_range=0.2,  # Shift images horizontally by 20%
    height_shift_range=0.2,  # Shift images vertically by 20%
    shear_range=0.2,  # Apply shearing
    zoom_range=0.2,  # Random zoom in/out
    horizontal_flip=True,  # Flip images horizontally
    brightness_range=[0.8, 1.2],  # Adjust brightness
    fill_mode='nearest'  # Fill missing pixels
)
```

```python
#image paths
neg_example = 'C:/Users/vella/Mastes Thesis/processed_data/no_pathology_train/f_0.jpg6cc12c80-c86c-4813-9971-280492b00017'
pos_example = 'C:/Users/vella/Mastes Thesis/processed_data/pathology_train/f_0_a.jpg0be0bd50-1173-4df0-bf2d-c9d82faf602a'
```

### Non Pathology Example

```python
img = image.load_img(neg_example)
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  

# generate augmented images
aug_iter = datagen.flow(img_array, batch_size=1)

# Visualise
fig, axes = plt.subplots(1, 6, figsize=(18, 5))

original_img = img_array[0].astype('uint8')  
axes[0].imshow(original_img)
axes[0].axis('off')
axes[0].set_title("Original Image", fontsize=14, fontweight='bold', color='blue')

# Display 5 augmented images
for i in range(5):
    augmented_image = aug_iter.next()[0]  # Get the next augmented image
    augmented_image = np.clip(augmented_image, 0, 255).astype('uint8')  # Ensure correct pixel values
    axes[i+1].imshow(augmented_image)
    axes[i+1].axis('off')
    axes[i+1].set_title(f"Augmented {i+1}", fontsize=14, fontweight='bold', color='green')

# Show the final plot
plt.tight_layout()
plt.show()
```
