# GLENDA Dataset Analysis

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Dataset Overview 

```python
# Data from the image
categories = ["Peritoneum", "Ovary", "Uterus", "Die", "No Pathology"]
values = [84, 27, 11, 29, 27]

# Colors using the Viridis colormap
colors = plt.cm.viridis([0.1, 0.3, 0.5, 0.7, 0.9])

# Create the pie chart
plt.figure(figsize=(8,6))
wedges, texts, autotexts = plt.pie(
    values, labels=categories, autopct="%1.1f%%", colors=colors, startangle=140, textprops={'fontsize': 12}
)

# Create a legend with actual numbers
legend_labels = [f"{category}: {value}" for category, value in zip(categories, values)]
plt.legend(wedges, legend_labels, title="Cases", loc="lower left", bbox_to_anchor=(1.2, 0.5), fontsize=12)

plt.title("Distribution of Cases by Class", fontsize=16)
plt.savefig('../images/deep_learning/glenda_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Pathology Dataset Overview 

```python
# Data extracted from the image
categories = ["Peritoneum", "Ovary", "Uterus", "DIE", "Total"]
cases = [84, 27, 11, 29, 102]
frames = [257, 53, 17, 55, 373]
annotations = [489, 54, 25, 59, 628]

x = np.arange(len(categories))  # x locations for the groups
width = 0.25  # Width of bars

# Use Viridis colormap
viridis = plt.get_cmap("viridis")

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width, cases, width, label="Cases", color=viridis(0.2))
bars2 = ax.bar(x, frames, width, label="Frames", color=viridis(0.5))
bars3 = ax.bar(x + width, annotations, width, label="Annotations", color=viridis(0.8))

# Annotate bars with values
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f"{height}",
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Offset for readability
                    textcoords="offset points",
                    ha="center", va="bottom", fontsize=10, color="black")

# Labels and formatting
ax.set_xlabel("Annotated Class", fontsize=12)
ax.set_ylabel("Count", fontsize=12)
ax.set_title("GLENDA Dataset Pathology Overview", fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=11)
ax.legend()

plt.xticks(rotation=20)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.savefig('../images/deep_learning/glenda_overview.png', dpi=300, bbox_inches='tight')
plt.show()
```
