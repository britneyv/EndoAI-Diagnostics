```python
import pandas as pd

# Load the CSV file
file_path = "clinical_model_results.csv"
df = pd.read_csv(file_path)

df = df.drop(['Unnamed: 0'], axis=1)

# Transpose the DataFrame while setting the first column as the index
df_transposed = df.set_index(df.columns[0]).T

# Convert numerical values to percentages with 2 decimal places
df_transposed = df_transposed.applymap(lambda x: round(x * 100, 2) if isinstance(x, (int, float)) else x)

# Display the transposed table
df_transposed
```
